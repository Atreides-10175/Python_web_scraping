## Функции для скрапинга

1 Функция извлечения текста по классу.

```python
def safe_extract(parent, class_name):
    """Извлекает текст из HTML-элемента по классу.
    
    Args:
        parent : Родительский HTML-элемент для поиска.
        class_name (str): CSS-класс искомого элемента.
    
    Returns:
        str|None: Текст элемента или None, если элемент не найден
    """
    elem = parent.find('div', class_=class_name)
    return elem.get_text(strip=True) if elem else None
```

2 Функция очистки Id карточки товара.

```python
def clean_card_id(card_id):
    """Очищает ID товара от служебных префиксов.
    
    Args:
        card_id (str): Исходный ID (например, 'bx_12345').
    
    Returns:
        str: Очищенный ID (например, '12345').
    """
    clean_id = card_id.replace("bx_", "")
    parts = card_id.split('_')
    return parts[-1] if len(parts) > 1 else clean_id
```

3 Функция обработки ценового значения

```python
def process_price(price_str):
    """Конвертирует строку с ценой в число (удаляет RUB символы).
    
    Args:
        price_str (str): Цена в формате "1 299 ₽".
    
    Returns:
        float|None: Числовое значение цены или None, если строка пуста.
    """
    if price_str:
        return float(price_str.replace('₽', '').replace(' ', ''))
    return None
```

4 Функция удаления блоков с карточками новинок(чтобы они не попадали в обработку)

```python
def remove_new_items(soup):
    """Удаляет HTML-блоки с новинками (карусели) из структуры страницы.
    
    Args:
        soup (bs4.BeautifulSoup): Объект парсера страницы
    """
    for carousel in soup.find_all('section', class_='carousel-block'):
        carousel.decompose()
```

5 Функция извлечения данных о книге из карточки

```python
def extract_book_data(item):
    """Извлекает структурированные данные о книге из HTML-карточки.
    
    Args:
        item: HTML-элемент карточки товара.
    
    Returns:
        dict: Словарь с ключами:
            - id (str)
            - author (str|None)
            - title (str|None)
            - new_price (float|None)
            - old_price (float|None)
    """
    card_id = item.get('id', 'no_id')
    new_price_str = safe_extract(item, 'product-card-price-true')
    return {
        'id': clean_card_id(card_id),
        'author': safe_extract(item, 'product-card-author'),
        'title': safe_extract(item, 'product-card-title'),
        'new_price': process_price(new_price_str),
        'old_price': process_price(safe_extract(item, 'product-card-price-old'))
    }
    
```

6 Функция обработки страницы по каждой книге

```python
def parse_book_page(url, headers):
    """Парсит HTML-страницу каталога книг, извлекая данные о каждой книге.
    
    Обрабатывает одну страницу, удаляет блоки с новинками, фильтрует карточки товаров
    и возвращает структурированные данные в виде списка словарей.

    Args:
        url (str): Полный URL страницы каталога.
        headers (dict): Заголовки для запроса (например, User-Agent).

    Returns:
        list[dict]: Список книг в формате:
                                        'id': str,           # Очищенный ID товара
                                        'title': str,        # Название книги
                                        'author': str,       # Автор (или None)
                                        'new_price': float,  # Актуальная цена
                                        'old_price': float   # Старая цена (или None)
                Пустой список при ошибках.
        ]
    """
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        for carousel in soup.find_all('section', class_='carousel-block'):
            carousel.decompose()
        books = []
        for item in soup.find_all('div', class_='product-card'):
            if item.find('span', class_='product-card-new'):
                continue  
            books.append({
                'id': clean_card_id(item.get('id', 'no_id')),
                'title': safe_extract(item, 'product-card-title'),
                'author': safe_extract(item, 'product-card-author'),
                'new_price': process_price(safe_extract(item, 'product-card-price-true')),
                'old_price': process_price(safe_extract(item, 'product-card-price-old'))
                        })
        return books
    except Exception as e:
        print(f"Ошибка парсинга страницы {url}: {e}")
        return []
```

7 Функция для сохранения данных в csv

```python
def save_to_csv(data, filename, intermediate=False):
    """
    Сохраняет данные в CSV
    Args:
        data: список словарей с данными
        filename: имя файла (без .csv)
        intermediate: если True, добавляет номер страницы к имени файла
    """
    if not data:
        print("\U000026A0 Нет данных для сохранения")
        return
    full_filename = f"{filename}_page_{len(data)}.csv" if intermediate else f"{filename}.csv"
    pd.DataFrame(data).to_csv(full_filename, index=False, encoding='utf-8-sig')
    if intermediate:
        print(f'\U0001F4BD Промежуточный результат сохранён в {full_filename}')
    else:
        print(f"\U00002705 Готово! Основной файл: {full_filename}")
```

8 Функция сбора данных со всех страниц(с параметрами промежуточного сохранения)

```python
def get_all_books(base_url, max_pages=5, save_interval=3):
    """Сбор данных со всех страниц с промежуточным сохранением.
    Для отключения промежуточного сохранения ввести save_interval > max_pages.
        
    Args:
        base_url: страница сайта для скрапинга
        max_pages: количество обрабатываемых страниц(по умолчанию 5)
        save_interval: интервал страниц для сохранения промежуточного результата(по умолчанию 3)
    """
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'}
    all_books = []
    
    for page in range(1, max_pages + 1):
        url = f"{base_url}/?PAGEN_1={page}" if page > 1 else base_url
        print(f"\n \U00002714  Парсинг страницы {page}...")
        books = parse_book_page(url, headers)
        if not books:
            print("\U0001F6A7 Достигнут конец каталога")
            break
        all_books.extend(books)
        if page % save_interval == 0:
            save_to_csv(all_books, 'books_catalog2', intermediate=True)
        delay = random.uniform(2, 5)
        print(f"\U0001F4A4 Задержка {delay:.1f} сек...")
        sleep(delay)
    save_to_csv(all_books, 'books_catalog2')
    return all_books
```

9 Запуск вэб-скрапинга

```python
if __name__ == '__main__':
    base_url = "https://moya-kniga.ru/catalog/otechestvennaya-8909"
    print("Начало парсинга...")
    all_books = get_all_books(
        base_url=base_url,
        max_pages=5,         
        save_interval=4       
    )
    if all_books:
        print(f"\n \U00002705 Итого собрано: {len(all_books)} книг")
    else:
        print("\U0001F6AB Не удалось получить данные")
```
